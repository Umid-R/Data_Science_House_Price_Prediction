```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../Datasets/cleaned_data.csv')
df.info()
```

```python

print(f"Number of total unique cities: {df['city'].nunique()}")
print(f"Average number of rooms: {df['bedrooms'].mean()}")
print(f"Year that the most houses were built: {df['yr_built'].mode()[0]}")
print(f"The most expensive house on the dataset: {df['price'].max()}")
print(f"The cheapest house on the dataset: {df['price'].min()}")


```

```python
# The most expensive houses for each city
df.groupby('city')['price'].max()
```

```python
# The cheapest houeses for each city
df.groupby('city')['price'].min()
```

```python
# Number of houses for each condition on the dataset
df['condition'].value_counts().sort_index()
```

```python
sns.set_theme(style="darkgrid")
sns.displot(
    df,
    x='price',
    color='steelblue',
    bins=60,
)

```

```python
sns.countplot(
    df,
    x='bedrooms'
)
```

```python
sns.countplot(
    df,
    x='condition'
)
```

```python
sns.countplot(
    df,
    x='floors'
)
```

```python
city_mean = df.groupby('city')['price'].mean()
df['city'] = df['city'].map(city_mean)
```

```python
df.to_csv('../Datasets/featured_data.csv', index=False)
```
